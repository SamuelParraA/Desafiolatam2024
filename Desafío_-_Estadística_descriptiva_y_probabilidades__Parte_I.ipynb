## Una comparación entre los salarios agrupados por 3 diferentes categorías a tu elección. ¿Para cuál de ellas las medidas de tendencia central son más representativas? ¿Para cuál son menos representativas? Explica.

```python
#Agrupamos por job_title y tomamos 3
grouped = salaries.groupby('job_title')['salary_in_usd'].describe()
grouped3=grouped.head(3)
grouped3
```

```python
# Calculamos las medidas de tendencia central
mean_salary = salaries.groupby('job_title')['salary_in_usd'].mean()
median_salary = salaries.groupby('job_title')['salary_in_usd'].median()
mode_salary = salaries.groupby('job_title')['salary_in_usd'].agg(lambda x: x.mode().values[0] if not x.mode().empty else None) #

# Para encontrar las más y menos representativas podríamos usar la desviación estándar
std_salary = salaries.groupby('job_title')['salary_in_usd'].std()

# Combinamos todo en un nuevo DataFrame summary
summary = pd.DataFrame({
    'Mean': mean_salary,
    'Median': median_salary,
    'Mode': mode_salary,
    'StdDev': std_salary
}).reset_index()

# Se seleccionan 3 categorías al azar

# Fijar la semilla para reproducibilidad
np.random.seed(123)

# Seleccionar 3 registros al azar
summary3 = summary.sample(n=3)

# Encuentra la medida más representativa (menor desviación estándar)
mas_representativa = summary3.loc[summary3['StdDev'] == summary3['StdDev'].min()]

# Encuentra la medida menos representativa (mayor desviación estándar)
menos_representativa = summary3.loc[summary3['StdDev'] == summary3['StdDev'].max()]

print("\nLa medida de tendencia central más representativa (menor desviación estándar):")
print(mas_representativa)

print("\nLa medida de tendencia central menos representativa (mayor desviación estándar):")
print(menos_representativa)
summary3.head(3)
```

 Por lo tanto de los 3 categorias la mas representativa sería Data Analytics Manager

## ¿Qué cargos reciben mejores sueldos en las empresas con sede en Estados Unidos?

```python

# Filtramos las empresas en Estados Unidos
us_based_companies = salaries[salaries['company_location'] == 'US']

# Agrupamos por título del trabajo y calculamos el salario promedio
average_salary_by_job = us_based_companies.groupby('job_title')['salary_in_usd'].mean()

# Ordenamos los resultados para obtener los trabajos mejor pagados
best_paid_jobs = average_salary_by_job.sort_values(ascending=False)

print("\nLos cargos mejor pagados en US son:")
best_paid_jobs.head(3)
```
